# Q.2. Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
imdb_url= "https://www.imdb.com/chart/top/"
page=requests.get(imdb_url)
page
```

```python
page.content
```

```python
soup= BeautifulSoup(page.content)
print(soup.prettify)
```

```python
movie = []
rating = []
yearofrelease = []
```

```python
for i in soup.find_all('td',class_="titleColumn"):
    
    movie.append(i.text.replace('\n'," "))
    if len(movie) == 50:
        break
movie
```

```python
for i in soup.find_all('td', class_="ratingColumn imdbRating"):
    
    rating.append(i.text.strip())
    if len(rating) == 50:
        break
rating
```

```python
for i in soup.find_all('span', class_="secondaryInfo"):
    yearofrelease.append(i.text.strip())
    if len(yearofrelease) == 50:
        break
yearofrelease
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'MoviesName':movie,'Ratings':rating,'Yearofrelease':yearofrelease})
df
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# 6) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame
# a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
page = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
soup = BeautifulSoup(page.content)
womens_team = soup.find_all("span",class_='u-hide-phablet')
womens_team_name = []
for i in womens_team:
    womens_team_name.append(i.text)
womens_matches = [] 
womens_points = [] 
womens_ratings = [] 
womens_new_list = [] 
for i in soup.find_all("td",class_='rankings-block__banner--matches'): 
    womens_matches.append(i.text)
for i in soup.find_all("td",class_='rankings-block__banner--points'):
    womens_points.append(i.text)
for i in soup.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    womens_ratings.append(i.text.replace("\n",""))
for i in soup.find_all("td",class_='table-body__cell u-center-text'):
    womens_new_list.append(i.text)
for i in range(0,len(womens_new_list)-1,2):
    womens_matches.append(womens_new_list[i]) 
    womens_points.append(womens_new_list[i+1])
for i in soup.find_all("td",class_='table-body__cell u-text-right rating'):
    womens_ratings.append(i.text)
womens_icc=pd.DataFrame({})
womens_icc['Team_name']=womens_team_name[:10]
womens_icc['Matches']=womens_matches[:10]
womens_icc['Points']=womens_points[:10]
womens_icc['Ratings']=womens_ratings[:10]
womens_icc    
```

# Q.7.  Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame
# i)Headline
# ii) Time
# iii) News Link

```python
page=requests.get("https://www.cnbc.com/world/?region=world")
page
```

```python
soup= BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
headlines=[]
for i in soup.find_all("a",class_="LatestNews-headline"):
    headlines.append(i.text)
headlines
```

```python
time=[]
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
time
```

```python
scraplink=soup.find_all('a',class_="LatestNews-headline")
scraplink
```

```python
links=[]
for i in scraplink:
    links.append(i.get("href").replace("\n"," ").strip(" "))
print(links)
```

```python
df_cnbc=pd.DataFrame({"Headline":headlines,"Time":time,"News Link":links})
df_cnbc
```

# 9) Write a python program to scrape mentioned details from dineout.co.in and make data frame
# i)Restaurant name
# ii) Cuisine
# iii) Location
# iv) Ratings
# v) Image URL

```python
url="https://www.dineout.co.in/delhi-restaurants/buffet-special"
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scraped_names=soup.find_all("div",class_="restnt-info cursor")
scraped_names
```

```python
names=[]
for i in scraped_names:
    names.append(i.get_text().replace("\n"," ").strip(" "))
names
```

```python
scraped_location=soup.find_all("div",class_="restnt-loc ellipsis")
scraped_location
```

```python
location=[]
for i in scraped_location:
    location.append(i.get_text().replace("\n"," ").strip(" "))
location
```

```python
scraped_cuisine=soup.find_all("span",class_="double-line-ellipsis")
scraped_cuisine
```

```python
cuisine=[]
for i in scraped_cuisine:
    cuisine.append(i.get_text().replace("\n"," ").strip(" "))
cuisine
```

```python
scraped_ratings=soup.find_all("div",class_="restnt-rating rating-4")
rating=[]
for i in scraped_ratings:
    rating.append(i.get_text().replace("\n"," ").strip(" "))
rating
```

```python
scraped_image=soup.find_all(class_="no-img")
scraped_image
```

```python
image=[]
for i in scraped_image:
    image.append(i.attrs['data-src'])
image
```

```python
df=pd.DataFrame()
df["Restaurant name"]=names
df["Cuisine"]=cuisine
df["Location"]=location
df["Ratings"]=rating
df["Image_URL"]=image
df
```
